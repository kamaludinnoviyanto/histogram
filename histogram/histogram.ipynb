```python
# import library

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load image

image = cv2.imread('ibra.jpg')
```

```python
# Convert from BGR to RGB

img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
print(img.shape)
```

```python
# histogram visualization

plt.hist(img.ravel(),256,[0,256]);
plt.show()
```

```python
# define the histogram -->> the higher the left-hand curve, the darker the image, and vice versa

color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()  
```
